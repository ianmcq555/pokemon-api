```python
def get_racer_info(data):
    new_data=[]
    for racer in data:
        racer_dict={}
        racer_name = f"{racer['Driver']['givenName']} {racer['Driver']['familyName']}"
        
        racer_dict[racer_name]={
            'first_name':racer['Driver']['givenName'],
            'last_name':racer['Driver']['familyName'],
            'position':racer['position'],
            'wins':racer['wins'],
            'DOB':racer['Driver']['dateOfBirth'],
            'nationality':racer['Driver']['nationality'],
            'contructor':racer['Constructors'][0]['name']
        }
        new_data.append(racer_dict)
    return new_data

get_racer_info(data)
```

```python
def getRacerInfo(year, match):
    url = f'http://ergast.com/api/f1/{year}/{match}/driverstandings.json'
    response = requests.get(url)
    if response.ok:
        my_dict = response.json()
        racer_dict = {}
        standings = my_dict["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
        for entry in standings:
            name = f"{entry['Driver']['givenName']}"
            racer_dict[name] = {
                'position': entry['position'],
                'team': entry['Constructors'][0]['name']
            }
        return racer_dict
    return 'Error... invalid year/match combo'

getRacerInfo(2020,1)
```

```python
def get_driver_info_by_year_and_round(year,r):
    url = f"http://ergast.com/api/f1/{year}/{r}/driverStandings.json"
    response = requests.get(url)
#     if not response.ok:
    if not response.json()["MRData"]["StandingsTable"]["StandingsLists"]:
        return "We had an error loading your data likely the year or round is not in the database"
    data = response.json()["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
    return get_racer_info(data)

get_driver_info_by_year_and_round(2020,1)
```

```python
def get_pokemon_info(pokemon):
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon}"
    response = requests.get(url)
    if response.ok:
        my_dict0 = response.json()
        pokemon_dict = {}
        info = my_dict0['name']['abilities'][0]['ability']['name']['base_experience']['sprites']['front_shiny']
        for entry in info:
            name = f"{entry['name']}"
            pokemon_dict[name] = {
                'name': entry['name'],
                'ability': entry['abilities'][0]['ability']['name'],
                'base_exp': entry['base_experience'],
                'image': entry['sprites']['front_shiny']
            }
        return pokemon_dict
    return 'Error... invalid pokemon'

print(get_pokemon_info('blastoise'))
print(get_pokemon_info('charizard'))
print(get_pokemon_info('venusaur'))
print(get_pokemon_info('pikachu'))
print(get_pokemon_info('mew'))
```
